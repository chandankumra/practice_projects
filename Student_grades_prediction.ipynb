```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=74)
```

```python
from sklearn.metrics import r2_score,auc,roc_auc_score,roc_curve
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
lr=LinearRegression()
knn=KNeighborsRegressor()
dtr=DecisionTreeRegressor()
rf=RandomForestRegressor()
```

```python
lr.fit(x_train,y_train)
print('linear_reg')
print('train_score',lr.score(x_train,y_train))
y_pred=lr.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
knn.fit(x_train,y_train)
print('KNN_reg')
print("train score",knn.score(x_train,y_train))
y_pred=knn.predict(x_test)
print('test score',r2_score(y_test,y_pred))
```

```python
dtr.fit(x_train,y_train)
print('decision tree')
print('train score',dtr.score(x_train,y_train))
y_pred=knn.predict(x_test)
print('test score',r2_score(y_test,y_pred))
```

```python
rf.fit(x_train,y_train)
print('random forest')
print('train_score',rf.score(x_train,y_train))
y_pred=rf.predict(x_test)
print('test_score',r2_score(y_test,y_pred))
```
